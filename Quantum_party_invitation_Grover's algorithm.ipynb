## *import qiskit library*

```python
from qiskit import BasicAer
from qiskit.aqua.algorithms import Grover
from qiskit.aqua.components.oracles import LogicalExpressionOracle
from qiskit.tools.visualization import plot_histogram
```

# *Creating a logical expression according to our conditions*
# *And implementing Grover's algorithm using the logical expression*

```python
# creating a logical expression according to our conditions
logical_expression = '((A & B) | (c & D)) & ~(B & D)'
# implementing Grover's algorithm using the logical expression 
algorithm = Grover(LogicalExpressionOracle(logical_expression))
```

## *A, B, C & D from the above cell might be compared to the below line practically*
## *logical_expression = '((Karthi & Arun) | (Sam & San)) & ~(Kush & Divya)"*

```python
# defining the QASM(Quantum Assembly Simulator)
backend = BasicAer.get_backend('qasm_simulator')
```

```python
# storing the output in result variable
result = algorithm.run(backend)
```

```python
plot_histogram(result['measurement'], title='Possible party combinations')
```
